```python
#Countour plots
import numpy as np
x = np.linspace(0,5,50)
y = np.linspace(0,5,50)
def func(x,y):
  return np.sin(x) ** 2 + np.cos(y)**2
X,Y = np.meshgrid(x,y)
Z=func(X,Y)
plt.contour(X,Y,Z,cmap='seismic')
```

```python
#backward fill - bfill command fills the null value with the previous value in the dataframe
#forward fill - ffill command fills the null value with the next value in the dataframe

df2 = df2.ID.fillna(method = 'ffill')
df2
```
